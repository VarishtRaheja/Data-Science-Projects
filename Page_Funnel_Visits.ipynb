```python
import pandas as pd
```

Import all_data the files

```python
visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
                   
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

Step 1: Inspect the DataFrames using `print` and `head`

```python
print(visits.head(5))
print(cart.head(5))
print(checkout.head(5))
print(purchase.head(5))
```

Step 2: Left merging visits and cart

```python
visits_cart_merge = pd.merge(visits,cart,how='left')
print(visits_cart_merge)
```

Step 3: How long is `visits_cart`?

```python
print(len(visits_cart_merge))
```

Step 4: How many timestamps are null for `cart_time`?

```python
cart_null_timestamps = len(visits_cart_merge[visits_cart_merge['cart_time'].isna()])
print(cart_null_timestamps)
```

Step 5: What percentage only visited?

```python
visited_not_cart = cart_null_timestamps/len(visits_cart_merge)
print(visited_not_cart)
```

Step 6: What percentage placed a t-shirt in their cart but did not checkout?

```python
checkout_cart = pd.merge(cart,checkout,how='left')
# print(checkout_cart)
no_checkout = checkout_cart.loc[checkout_cart['checkout_time'].isna()==True].count()
percentage_no_checkout = no_checkout.user_id/len(checkout_cart)
print(f"{percentage_no_checkout:.2%}")
```

Step 7: Merge it all_data together

```python
all_data = visits.merge(cart,how='left').merge(checkout,how='left').merge(purchase,how='left')
all_data
```

Step 8: % of users who got to checkout but did not purchase

```python
# group = all_data.groupby(['checkout_time', 'purchase_time'], dropna=False)['user_id'].count().reset_index()
# print(group)
users_in_checkout = all_data[all_data['checkout_time'].isna()==False]
users_no_purchase = all_data[(all_data['purchase_time'].isna()==True) & (~all_data['checkout_time'].isna())]
checkout_not_purchase_percent = len(users_no_purchase)/len(users_in_checkout)
print(checkout_not_purchase_percent)
```

Step 9: check each part of the funnel, let's print all_data 3 of them again

```python
print("{} percent of users who visited the page did not add a t-shirt to their cart".format(round(visited_not_cart*100, 2)))
print("{} percent of users who added a t-shirt to their cart did not checkout".format(round(percentage_no_checkout*100, 2)))
print("{} percent of users who made it to checkout did not purchase a shirt".format(round(checkout_not_purchase_percent*100, 2)))
```

*The weakest part of the funnel is clearly getting a person who visited the site to add a tshirt to their cart. Once they've added a t-shirt to their cart it is fairly likely they end up purchasing it. A suggestion could be to make the add-to-cart button more prominent on the front page.*


Step 10: adding new column

#Using the giant merged DataFrame all_data that you created, let’s calculate the average time from initial visit to final purchase. Add a column that is the difference between purchase_time and visit_time.

```python
all_data['diff_purchase_visit'] = all_data['purchase_time'] - all_data['visit_time']
```

Step 11: examine the results

```python
print(all_data['diff_purchase_visit'])
```

Step 12: average time to purchase

```python
avg_time = all_data['diff_purchase_visit'].mean()
print(avg_time)
```
